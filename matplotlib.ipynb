```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Line graph using Matplotlib and Numpy

```python
a=np.array([1,4,9,16,25])
b=np.array([1,2,3,4,5])
c=np.array([25,20,15,10,5])
plt.figure(figsize=(10,5))
plt.plot(a,'bo-',linewidth=2.,label='line1')
plt.plot(b, 'r+-', label='line 2', linewidth=2)
plt.plot(c,'b^-',label='line-3')
plt.legend(loc="upper left")
plt.title('Sample-01',fontsize=20,fontweight='bold')
plt.xticks(range(0,len(a)),('jan','feb','march','april','may'))
plt.xlabel("Months")
plt.ylabel("Production")
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("/content/Marketing Raw Data.csv")
df.head(5)
```

```python
# SIMPLE LINE GRAPH
import matplotlib.pyplot as plt
plt.plot(df['Visitors'])
plt.show()
```

```python
a=df.tail(10).reset_index()
plt.figure(figsize=(5,5))
plt.title('Visitors vs Revenue',fontsize=20,fontweight='bold')
plt.plot(a['Visitors'],'bo-',label='visitors')
plt.plot(a['Revenue'],'r^-',label='Revenue')
plt.xticks(range(0,len(a['Visitors'])),a['Date'],rotation=20)
plt.legend(loc="upper right",fontsize=12)
plt.xlabel('Dates attended by visitors')
plt.ylabel("visitors")
plt.show()
```

```python

plt.figure(figsize=(10,10))
plt.title('Visitors vs Revenue',fontsize=20,fontweight='bold')
plt.bar(range(0,len(a['Visitors'])),a['Visitors'],label='visitors')
plt.bar(range(0,len(a['Marketing Spend'])),a['Marketing Spend'],label='marketing cost')
plt.plot(a['Revenue'],'r^-',label='Revenue')
plt.xticks(range(0,len(a['Visitors'])),a['Date'],rotation=20)
plt.legend(loc="upper right",fontsize=12)
plt.xlabel('Dates attended by visitors')
plt.ylabel("visitors")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Visitors vs Revenue',fontsize=20,fontweight='bold')
plt.bar(range(0,len(a['Revenue'])),a['Revenue'],label='revenue')
plt.bar(range(0,len(a['Visitors'])),a['Visitors'],label='visitors')

plt.plot(a['Revenue'],'r^-',label='Revenue')
plt.xticks(range(0,len(a['Visitors'])),a['Date'],rotation=20)
plt.legend(loc="upper right",fontsize=12)
plt.xlabel('Dates attended by visitors')
plt.ylabel("visitors")
plt.show()
```

```python
a[['Revenue','Marketing Spend']].plot(kind='bar',color=['blue','red'],figsize=(10,5))
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(5,5))
plt.scatter(df['Revenue'],df['Marketing Spend'],c='red',marker='*',label="Revenue")
plt.title("Revenue vs Marketing",fontsize=20,fontweight='bold')
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")
plt.legend(loc="upper center")
```

```python
#Scatter plot groups
group=df[["Promo","Revenue","Marketing Spend"]].groupby("Promo")

fig,ax=plt.subplots(figsize=(15,10))
for promo,group in group:
  ax.scatter(group["Revenue"],group["Marketing Spend"],marker='*',label=promo)
ax.legend("upper left")
plt.show()
```
