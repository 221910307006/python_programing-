```python
import cv2
import  pandas as pd
import numpy as np
```

```python
data=cv2.imread('/content/fresh-fruits.webp')
print(type(data))
```

```python
print(data.shape) # 280,383 is the resolution 3 is RGB
```

```python
import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(data, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
#Gray Scale Image
data1=cv2.cvtColor(data,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(data1)
plt.show()
```

```python
print(data1.shape) # no RGB only resolution
```

```python
#Resize an image
print(data.shape)
img_resize=cv2.resize(data,(data.shape[0]//2, data.shape[1]//2))
print(img_resize.shape)
```

```python
#Flipping an image
tri=cv2.imread('/content/tri.png')
img_flip=cv2.flip(tri,0) # x axis flip if it is zero or y-axis flip
plt.imshow(img_flip)
plt.axis('off')
plt.show()
```

```python

img_flip=cv2.flip(data,1) # x axis flip if it is zero or y-axis flip
plt.imshow(img_flip)
plt.axis('off')
plt.show()
```

```python
img_crop=data[0:300,0:500]
plt.imshow(img_crop)
#plt.axis('off')
plt.show()
```

```python
# Drawing shapes
slate=np.zeros((512,512,3))

cv2.rectangle(slate,pt1=(100,100),pt2=(150,200),color=(0,255,0),thickness=6)

cv2.circle(slate,center=(300,300),radius=40,color=(250, 2, 209),thickness=-1)

cv2.line(slate,pt1=(100,100),pt2=(200,200),color=(0,0,255),thickness=6)

cv2.putText(slate,org=(100,100),fontScale=4,color=(0,255,255),thickness=2,lineType=cv2.LINE_AA,text="Hello..",fontFace=cv2.FONT_ITALIC)
plt.imshow(slate)
```
