```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("/content/seaborn_excel.csv")
#df.head(5)
df.tail(5)
```

###Line Graph###

```python
#Simple line graph
#Assuming we want to invest the revenue by date

ax=sns.lineplot(x="Week_ID",y="Revenue",data=df)
```

```python
# Hue: It is the key word used to split the data into 3 types ie No promo,Promotion Red,Promotion Blue
# ci: removes continuous intervals when it is set to false
# marker: It is used to spot/dot the pts
ax=sns.lineplot(x="Week_ID",y="Revenue",data=df,hue="Promo",style='Promo',ci=False,markers=True)
sns.set(rc={'figure.figsize':(12,10)})
```

```python
#To see the data by mounth
ax=sns.lineplot(x="Month_ID",y="Revenue",data=df,hue="Promo",style="Promo",ci=False,markers=True)
```

```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
df1=pd.read_csv('/seaborn_excel.csv')
ax=sns.boxplot(x='Day_Name',y='Revenue',data=df1,color='#36c753')
ax=sns.swarmplot(x='Day_Name',y='Revenue',data=df1,color='#9cd6d9')
```

```python
df1.columns
```

```python
ax=sns.scatterplot(x='Marketing Spend',y='Revenue',data=df1,hue='Promo',style='Promo',size='Revenue')
```

###*LM PLots*

```python
ax=sns.lmplot(x='Marketing Spend',y='Revenue',data=df1,size=5)
```

```python
ax=sns.lmplot(x='Marketing Spend',y='Revenue',data=df1,hue='Promo',height=5,ci=None,markers=['*','+','o'])
```

```python
fig,size=plt.subplots(2,2,figsize=(10,10))
a1=df1['Revenue'].values
a2=df1['Marketing Spend'].values
a3=df1['Visitors'].values
# plotted at 1,0 
sns.distplot(a1,color='red',ax=size[0,0])

sns.distplot(a2,color='red',ax=size[0,1])

sns.distplot(a3,color='red',ax=size[1,0])

ax=sns.boxplot(x='Day_Name',y='Revenue',data=df1,color='#36c753')
ax=sns.swarmplot(x='Day_Name',y='Revenue',data=df1,color='#9cd6d9')
```

```python
# Pair plot
g=sns.pairplot(df1[['Revenue','Marketing Spend','Visitors','Promo']],plot_kws={'color':'green'})
```

```python
# Pair plot
g=sns.pairplot(df1[['Revenue','Marketing Spend','Visitors','Promo']],plot_kws={'color':'green'},hue='Promo')
```

```python
# Pair plot
g=sns.pairplot(df1[['Revenue','Marketing Spend','Visitors','Promo']],plot_kws={'color':'green'},hue='Promo',kind='reg')
```

###*Joint plot*

```python
sns.jointplot('Marketing Spend','Revenue',data=df1,color='blue',height=5,kind='reg')
```

```python
col=['Marketing PC','Revenue PC','Visitors PC']
c1=df1[['Marketing Spend','Revenue','Visitors']].corr()
print(c1)
hm=sns.heatmap(c1,annot=True,
               annot_kws={'size':'20'},xticklabels=col,yticklabels=col,cmap='BuPu')
```
